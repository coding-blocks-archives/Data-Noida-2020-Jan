```python
class Human:
    def __init__(self, name="Unnamed", passwd="password"):
        self.__name = name
        self.__age = 0
        self.__pass = passwd
        
    def getName(self):
        return self.__name
    
    def setName(self, name):
        self.__name = name 
        
        
    def getAge(self):
        return self.__age
    
    def setAge(self, age):
        password = input("Enter password")
        if password != self.__pass:
            return
        
        if age in range(self.__age, self.__age+2):
            self.__age = age
            
```

```python
class Human:
    
    population = 0
    listing = []
    
    def __init__(self, name="Unnamed"):
        self.name = name
        self.alive = True
    
        Human.population += 1
        
        Human.listing.append(self)
        
    @classmethod
    def freeKill(cls, people, target):
        if sum([item.alive for item in Human.listing]) >= 3:
            target.die()
        
    def die(self):
        Human.population -= 1
        self.alive = False
        
    def __repr__(self):
        return "{} is {}".format(self.name, self.alive)
```

```python
class Hitman(Human):
    
    def __init__(self, name="Hitler"):
        super().__init__(name)
        
        self.kills = 0
    
    def shoot(self, person):
        if person == self:
            print("Suicide is never a way")
        elif self.alive and person.alive:
            person.die()
            self.kills += 1
            
    def __repr__(self):
        return "{} is {} and killed {}".format(self.name, self.alive, self.kills)
```
